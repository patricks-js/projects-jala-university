### **2 - Estrutura de um Método**
- Palavra reservada **SELF**:

    - O **Self** é usado em classes no Python para indicar que você está referenciando alguma coisa do próprio objeto (sejam eles atributos ou métodos); na verdade podemos afirmar que ***o self é o próprio objeto em si.***


---


- Palavra reservada **__ init __**:

    - O **__ init __** é usado para inicializar o estado do objeto. A tarefa dos construtores é inicializar (atribuir valores) aos membros de dados da classe quando um objeto da classe é criado.
      - Um exemplo pratico, seria se nós criássemos um objeto chamado Água, sabemos que o init da agua seria (Transparente, sem cheiro, sem gosto e no estado liquido).


---


- Palavra reservada **Print**:

    - O **Print** é usado para que o programa imprima na tela, o que você determinou, podendo ser algo que você escreveu para guiar o usuário ou até mesmo, variáveis ou classes.

### **3 - Boas práticas**
- Uso do **camelCase**:

    - Essa nomenclatura foi criada por desenvolvedores para deixar a leitura dos códigos mais compreensíveis, exemplo:
      - Sem camelCase: notadoaluno
      - Com camelCase: notaDoAluno

---

- Uso da identação:
  - A identação é empregada com o objetivo de ressaltar a estrutura do algoritmo, aumentando assim a legibilidade do código.
    - Código Python com identação:

```python
# Código Python com identação:
valor = 1
if valor > 1:
  if valor > 7:
    print("valor alto")
else:
  print("valor igual a", valor)
```

```python
# Código Python sem identação:
valor = 1
if valor > 1:
if valor > 7:
print("valor alto")
else:
print("valor igual a", valor)
```

## **Código para implementar no projeto final do Robô**

- Código desenvolvido com o conhecimento e exemplos passados durante a aula dois.

```python
class Part():
    
    def __init__(self, name: str, attack_level=0, defense_level=0, energy_consumption=0):
        self.name = name
        self.attack_level = attack_level
        self.defense_level = defense_level
        self.energy_consumption = energy_consumption
        
    def get_status_dict(self):
        formatted_name = self.name.replace(" ", "_").lower()
        return {
            "{}_name".format(formatted_name): self.name.upper(),
            "{}_status".format(formatted_name): self.is_available(),
            "{}_attack".format(formatted_name): self.attack_level,
            "{}_defense".format(formatted_name): self.defense_level,
            "{}_energy_consump".format(formatted_name): self.energy_consumption,
        }
        
    def reduce_defense(self, attack_level):
        self.defense_level = self.defense_level - attack_level
        if self.defense_level <= 0:
            self.defense_level = 0

    def is_available(self):
        return self.defense_level <= 0

class Robot:
    def __init__(self, name, color_code):
        self.name = name
        self.color_code = color_code
        self.energy = 100
        self.parts = [
            Part("Head", attack_level=5, defense_level=10, energy_consumption=5),
            Part("Weapon", attack_level=15, defense_level=0, energy_consumption=10),
            Part("Left Arm", attack_level=3, defense_level=20, energy_consumption=10),
            Part("Right Arm", attack_level=6, defense_level=20, energy_consumption=10),
            Part("Left Leg", attack_level=4, defense_level=20, energy_consumption=15),
            Part("Right Leg", attack_level=8, defense_level=20, energy_consumption=15),
        ]
        def greet(self):
            print("Hello, my name is", self.name)
    
    
        def print_energy(self):
            print("We have", self.energy, " percent energy left")
```
