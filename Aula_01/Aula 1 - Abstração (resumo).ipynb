### **1 - Entendendo abstração**
- O que é abstração?

     A Abstração seria  a extração das características que tornam um objeto único, em outras palavras, seria as características que distinguem um objeto de outros objetos.

- Visão externa de um objeto
  - O que todo objeto *tem*
  - O que todo objeto sabe *fazer*


      

### **2 - Entendendo as classes**
- Exemplo prático:

  - Classe: Gato
  - O que todo gato tem:
     - Atributos:
      - Nome
      - Tipo de pelo
      - Olhos

  - O que todo gato sabe fazer:
     - Métodos:
      - Miar
      - Comer
      - Dormir


---

- Criando uma instância:

    - Gato Branco
      - Nome: Lucius
      - Tipo de pelo: Branco
      - Olhos: Dourados

    - Gato Preto
      - Nome: Morpheus
      - Tipo de pelo: Preto
      - Olhos: Verdes


---

- Classe: 
  - É uma abstração de um conjunto de objetos.
- Atributo:
  - São características que todo objeto tem.
- Métodos:
  - São operações e ações que todo objeto sabe fazer.
- Objetos:
  - São instâncias específicas de uma classe.


- Anteriormente foi criada uma classe chamada Gato, que possuía seus atributos (Nome, Tipo do pelo e olhos) e seus métodos (Miar, Comer e Dormir). A partir dessa classe foram criados dois objetos o Gato Branco e o Gato Preto.

### **2 - Registro na Polygoland**
- Exemplo prático:

  - Registro na Polygoland. Polygoland é uma cidade composta apenas por cidadãos que são polígonos regulares. A prefeitura quer que todos sejam cadastrados, mas eles têm um problema... não sabem quais dados cadastrar, pois nem todos os polígonos são exatamente iguais.

O prefeito deu-lhe a tarefa de criar um modelo que represente todos os cidadãos de uma forma geral, as características que têm e o que podem fazer.

  - Um polígono regular é definido quando tem todos lados iguais, seja interno ou externo.
  - Para verificar se é um poligono regular ou não calculamos o perímetro (P = s * n) e a área (A = (P * a)/2)

- Criaremos a seguir uma classe para o problema dado acima:
      Classe Poligono
        Atributo
          - Side
          - Number of Side
          - Apothem
        Métodos
          - Calculate Perimeter
          - Calculate Area
