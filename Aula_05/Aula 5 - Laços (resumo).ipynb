## **Parte teórica**

- O que é um Laço de Repetição?
  
  As linguagens de programação normalmente suportam tanto estruturas condicionais como também os chamados laços de repetição, estruturas que permitem a execução de instruções repetidas vezes, até que uma condição seja atingida.

- Operadores Relacionais:
  - Como usar o For?
     
     O FOR nos permite percorrer os itens de uma coleção e, para cada um deles, executar um bloco de código. Como no exemplo abaixo:
     
     image.png
    
  - Como usar o While?
     
     O WHILE, executa um conjunto de instruções várias vezes enquanto uma condição é atendida.

     image.png
    
  

## **Parte prática**

```python
condicion = True

while condicion:
    print("true")
    condicion = False
```

```python
elemento = [1,2,3]

for variable in elemento:
    print(variable)
```
